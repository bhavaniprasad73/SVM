```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import LinearSVC
import sklearn as svm
input = pd.read_csv('Voting.txt', header=None)
print(input.shape)
print(type(input))
print("\n Original Data ")
print(input.head())

# 1. a) Convert this dataset into numeric by converting y to 1, n to -1, and ? to 0.
# input[1] = input[1].str.lower().replace({'yes': 1, 'no': 0})
input.replace(['y','n', '?'],[1,-1, 0],inplace=True)
print("\n 1. a) After converting this dataset into numeric by converting y to 1, n to -1, and ? to 0. \n")
print(input.head())

print(list(input))
list_labels = input[0].tolist()
print(list_labels)
plt.hist(list_labels)
 
your_bins= 3
data=[]
arr = plt.hist(list_labels,bins=your_bins)
for i in range(your_bins):
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]))
plt.xlabel('Class')
plt.ylabel('Count')

plt.title(r'Histogram for Classes')

plt.show() 
count_classes = input.groupby(0).size()

print(count_classes)
democrat = count_classes[0]
republican = count_classes[1]
print('democrat ', democrat, 'republican ', republican, 'ratio ', republican/democrat)
```

```python
from sklearn.model_selection import StratifiedKFold
import numpy as np
class1 = input[0] == 'democrat'
df_class1 = input[class1]
df_class2 = input[~class1]
#print(df_class1)
#print(df_class2)

part = df_class1.sample(frac=0.25, replace=False)
part1 = np.array_split(df_class1, 4)
demo_class1 = part1[0]
demo_class2 = part1[1]
demo_class3 = part1[2]
demo_class4 = part1[3]
print('\n demo 1', demo_class1)
print('demo 2', demo_class2)
print('demo 3', demo_class3)
print('demo 4', demo_class4)

part2 = np.array_split(df_class2, 4)
repub_class1 = part2[0]
repub_class2 = part2[1]
repub_class3 = part2[2]
repub_class4 = part2[3]
print('\n repub 1', repub_class1)
print('repub 2', repub_class2)
print('repub 3', repub_class3)
print('repub 4', repub_class4)

print('shape', repub_class1.shape)

frame1 = [demo_class1, repub_class1]
frame2 = [demo_class2, repub_class2]
frame3 = [demo_class3, repub_class3]
frame4 = [demo_class4, repub_class4]

fold1 = pd.concat(frame1)
fold2 = pd.concat(frame2)
fold3 = pd.concat(frame3)
fold4 = pd.concat(frame4)

print('fold1')
print( fold1.shape)


Cross_validation_fold = fold3.iloc[:, 1:]

Cross_validation_fold_label = fold3.iloc[:, 0]
print('Cross 12', Cross_validation_fold.head())

training12 = fold1.append(fold2, ignore_index=True)
training12_fold = training12.iloc[:, 1:]
training12_label = training12.iloc[:, 0]
print('train 12 ', training12_fold.head())

training24 = fold2.append(fold4, ignore_index=True)
training24_fold = training24.iloc[:, 1:]
training24_label = training24.iloc[:, 0]
training41 = fold4.append(fold1, ignore_index=True)

training41_fold = training41.iloc[:, 1:]
training41_label = training41.iloc[:, 0]
print('label\n', training41_label)

print(training12_fold)

test12_fold = fold4.iloc[:, 1:]

test12_fold_label = fold4.iloc[:, 0]

test24_fold = fold1.iloc[:, 1:]

test24_fold_label = fold1.iloc[:, 0]

test41_fold = fold2.iloc[:, 1:]

test41_fold_label = fold2.iloc[:, 0]
```

```python
#  Parameter tuning step optimal C for Linear SVM

from sklearn import svm
pred_list12 = []
c = 0.1
c_list = []
while c < 100:
    clf = svm.SVC(kernel = 'linear', C = c)
    clf.fit(training12_fold, training12_label)
    pred12 = clf.predict(Cross_validation_fold)
     # print(clf.predict(Cross_validation_fold))
    from sklearn.metrics import accuracy_score
    # print('accuracy')
    accuracy12 = accuracy_score(Cross_validation_fold_label, pred12)
    # print(accuracy12)
    pred_list12.append(accuracy12)
    
    c_list.append(c)
    c = c + 0.1
    
print('List')   
print(pred_list12)
    
l2 = [pred_list12.index(x) for x in sorted(pred_list12)]
print('index')
position12 = l2[-1]
print(position12)    
Optimal12 =  c_list[position12]    
print('c list ', c_list)    


print('Best Parameter C for Linear SVM is ', Optimal12)





```

```python
# Testing set accuracy
clf12 = svm.SVC(kernel = 'linear', C = Optimal12)
clf12.fit(training12_fold, training12_label)
pred12 = clf12.predict(Cross_validation_fold)
     # print(clf.predict(Cross_validation_fold))
from sklearn.metrics import accuracy_score
    # print('accuracy')
accuracy12 = accuracy_score(Cross_validation_fold_label, pred12)
    # print(accuracy12)
pred_list12.append(accuracy12)



pred12_test = clf12.predict(test12_fold)

print(pred12_test)

test_accuracy12 = accuracy_score(test12_fold_label, pred12_test)
Test_accuracy12 = 100 * test_accuracy12
print('\nAccuracy on testing set 12 : ', Test_accuracy12 , '%')

```
